```python
import pandas as pd
import os

# Base directory (project root)
BASE_DIR = os.path.dirname(os.path.abspath(__file__))

# Paths
data_folder = os.path.join(BASE_DIR, "Backend", "Data")

# ----------- EXPERIMENT DATA -----------
experiment_csv = os.path.join(data_folder, "experiment_brand_diverse.csv")
experiment_pkl = os.path.join(data_folder, "experiment_runtime.pkl")

print("Loading experimental CSV...")
df_exp = pd.read_csv(experiment_csv)

print("Saving experimental Pickle...")
df_exp.to_pickle(experiment_pkl)

print("Experimental dataset converted successfully âœ…")


# ----------- GEOMETRY DATA -----------
geometry_csv = os.path.join(data_folder, "geometry_brand_diverse.csv")
geometry_pkl = os.path.join(data_folder, "geometry_runtime.pkl")

print("Loading geometry CSV...")
df_geo = pd.read_csv(geometry_csv)

print("Saving geometry Pickle...")
df_geo.to_pickle(geometry_pkl)

print("Geometry dataset converted successfully âœ…")

print("All conversions completed ðŸš€")
```
